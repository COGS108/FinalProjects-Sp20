### Sentiment Analysis  

### Seting up the notebook and data for sentiment analysis

```python
allSchools = pd.read_csv('Other Notebooks/allSchools.csv')
```

```python


# Cleaning up the 'content' column by removing incidents of '\r', '\n' and, '\n\n'
allSchools['content'] = allSchools['content'].apply(lambda x: x.replace('\n',' ')) 
allSchools['content'] = allSchools['content'].apply(lambda x: x.replace('\r',' '))
allSchools['content'] = allSchools['content'].apply(lambda x: x.replace('\n\n',' '))
```

### Modifying the content column for sentiment analysis

```python
# tokenizing the content of the update
allSchools['token'] = allSchools['content'].apply(word_tokenize)

# remove stop words
allSchools['stop'] = allSchools['token'].apply(lambda x: [item for item in x if item not in stop_words])

# stemming
allSchools['stem'] = allSchools['stop'].apply(lambda x: [ps.stem(y) for y in x])

# get series of all words after stemming
stem_allSchools = allSchools['stem'].apply(pd.Series).stack()

# calculation word frequency
fdist_allSchools = FreqDist(stem_allSchools)

# remove punctuation counts
for punc in string.punctuation:
    del fdist_allSchools[punc]
    
# get list of the 'sentences' (responses) from each individual
allSchools_list = list(allSchools['content'].values)
```

## allSchools got an update!  
Now contains 3 new columns  
- **token**: Tokenization is the first step in sentiment analysis
    - By breaking down the text (tokenize) into smaller pieces (words, sentences, bigrams, etc.)
    - A token is a single entity
- **stop**: Stop words are of general disinterest in sentiment analysis.
    - Words such as, this, a, an, as, the, etc. do not hold very much importance.
    - Removal of stop words from our analysis will be a lower computational burden.
- **stem**: 
    - Now that we have tokenized our words and removed stop words, we can stem.
    - Stemming normalizes occurrence by reducing words to their root and removing word-endings such as, 'ing,' or 'ed.'

### Get the sentiment measures 
*using a get_sentiments function defined in functions.py*

```python
def get_sentiments(input_list):
    
    output = pd.DataFrame()

    for sentence in input_list:
        ss = analyser.polarity_scores(sentence)
        ss['sentence'] = sentence
        output = output.append(ss, ignore_index=True)

    return output
```

```python
analyser = SentimentIntensityAnalyzer()

# get sentiment measures
allSchools_sentiments = get_sentiments(allSchools_list)

allSchools = allSchools.drop(['token', 'stop'], axis = 1)\
.assign(pos = allSchools_sentiments['pos'])\
.assign(neg = allSchools_sentiments['neg'])\
.assign(neu = allSchools_sentiments['neu'])\
.assign(compound = allSchools_sentiments['compound'])

#allShools.to_csv('allShools_sent.csv', index = False)

#print the dataframe
allSchools.head()
```

### Now our table is complete now!
With 4 new columns 
- **pos**: The estimated positive sentiment of the given update
- **neg**: The estimated negative sentiment of the given update
- **neu**: The estimated neutral sentiment of the given update
- **compound**: Worksto combine all these ratings in a single number. 
    - \> .05 is positive
    - < -.05 is negative
    - in between is neutral

### Function to filter the sentiments by a key word

```python
def sort_df(word):
    new = allSchools[allSchools['content'].str.contains(word, case=False)]
    new = new.reset_index(drop = True)
    new_list = list(new['content'].values)
    new_sentiments = get_sentiments(new_list)
    
    new = new.assign(pos = new_sentiments['pos'])
    new = new.assign(neg = new_sentiments['neg'])
    new = new.assign(neu = new_sentiments['neu'])
    new = new.assign(compound = new_sentiments['compound'])
    
    return new
```
