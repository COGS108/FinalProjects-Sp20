# Background and Prior Work

Before discussing the background of our topic, we need to first understand the following sports betting terms:
  -  Bookmaker (bookies): The sports betting operators
  -  Money Line: 
     -  If the team bet on is the favourite, then the money line represents the amount of money a person must wage in order to profit `$`100. 
     -  If the team bet on is the underdog, then the money line represents the amount of money a person will profit when wagering `$`100.
  -  Spread: The spread represents how much the favourite must win by in order to profit.
  -  Opening/Closing: 
     -  Opening Money Line/Spread represents the Bookmakers’ predictions on the expected winner.
     - Closing Money Line/Spread represents the initial Bookmakers’ predictions on the expected winner with the influence of the public’s bets. Note, we simplify the closing prediction to represent the public’s opinion.

# Data Cleaning

Next, we have to import the data for each respective sport so that we can analyze it.  For each sport, the data is in an excel file with three sheets.  In all of the files, the last two sheets are empty, so we only import the first sheet into each sport dataframe. We stripped the excel file to get our data into a usable format and used the `read_excel` function to read the data into a dataframe.

Note: The Season_Id column contains the starting year of the season because seasons usually span multiple years.

```python
def import_basketball_data():
    basketball_df = pd.DataFrame()
    
    files = glob.glob('./basketball_data/*.xlsx')
    for file in files:
        striped_file = file.strip()
        starting_year = striped_file[27:34].split('-')[0]
        df = pd.read_excel(file.strip(), sheet_name='Sheet1')
        df['Season_Id'] = int(starting_year)
        basketball_df = basketball_df.append(df)
        
    return basketball_df
```

### Group the games
In the original dataframe for each sport, every two columns represents one match for the two teams in each row.  To simplify analysis, we decided to group these rows so that each row is a match between two teams.  For basketball and football, the spread (which is an important predictor for the winner) is in the same column as the over-under, which does not tell us anything about the potential winner.  In each match, the over-under is always larger than the spread, so to isolate the spread we take the larger value in each match and set the lesser value to `np.nan`. In baseball and hockey, the predictions are made with a money line, so we do not have this issue.

```python
def group_games(df):
    new_df = pd.DataFrame()
    for i in range(0,len(df), 2):
        first_team = df.loc[i, 'Team']
        second_team = df.loc[i+1, 'Team']
        
        first_score = df.loc[i, 'Final']
        second_score = df.loc[i+1, 'Final']
        
        if first_score == 'NL' or second_score == 'NL':
            continue
        
        combined_score = str(first_score) + '-' + str(second_score)
        
        season = df.loc[i, 'Season_Id']
        
        if 'ML' in df.columns:
            first_ML = df.loc[i, 'ML']
            second_ML = df.loc[i+1, 'ML']
            
            first_open = df.loc[i, 'Open']
            second_open = df.loc[i+1, 'Open']
            
            if first_open == 'pk' or first_open == 'PK':
                first_open = 0
                second_open = np.nan
            elif second_open == 'pk' or second_open == 'PK':
                second_open = 0
                first_open = np.nan
            elif int(first_open) < int(second_open):
                second_open = np.nan
            else:
                first_open = np.nan

            first_close = df.loc[i, 'Close']
            second_close = df.loc[i+1, 'Close']
            
            if first_close == 'pk' or first_close == 'PK':
                first_close = 0
                second_close = np.nan
            elif second_close == 'pk' or second_close == 'PK':
                second_close = 0
                first_close = np.nan
            elif int(first_close) < int(second_close):
                second_close = np.nan
            else:
                first_close = np.nan

            
            new_df = new_df.append({'first_team' : first_team, 'second_team' : second_team, 
                                'score' : combined_score, 'first_open_spread' : first_open,
                                'second_open_spread' : second_open, 'first_close_spread' : first_close,
                                'second_close_spread' : second_close, 'first_ML' : first_ML,
                                'second_ML' : second_ML, 'season' : season}, ignore_index=True)
        else:
            first_open = df.loc[i, 'Open']
            second_open = df.loc[i+1, 'Open']

            first_close = df.loc[i, 'Close']
            second_close = df.loc[i+1, 'Close']
            
            if 'Run Line' in df.columns:
                first_run_line = df.loc[i, 'Run Line']
                second_run_line = df.loc[i+1, 'Run Line']
                new_df = new_df.append({'first_team' : first_team, 'second_team' : second_team, 
                                'score' : combined_score, 'first_open' : first_open,
                                'second_open' : second_open, 'first_close' : first_close,
                                'second_close' : second_close, 'first_run_line' : first_run_line,
                                'second_run_line' : second_run_line, 'season' : season}, ignore_index=True)
            else:
                first_puck_line = df.loc[i, 'Puck Line']
                second_puck_line = df.loc[i+1, 'Puck Line']
                new_df = new_df.append({'first_team' : first_team, 'second_team' : second_team, 
                                'score' : combined_score, 'first_open' : first_open,
                                'second_open' : second_open, 'first_close' : first_close,
                                'second_close' : second_close, 'first_puck_line' : first_puck_line,
                                'second_puck_line' : second_puck_line, 'season' : season}, ignore_index=True)

    new_df = new_df.astype({'season': 'int32'})
    return new_df
```

### Find Match Favorite

Here, we determine the match favorite using the predictor, whether it be a money line or spread.  
- If it's a spread, then whichever team the spread is attached to is the favorite (the other team should already be `NaN`).
- If it's an over under, whichever team has a lesser value is the favorite.  

```python
def find_favorites(df):
    series_open = []
    series_close = []
    for i in range(0, len(df)):
        if 'first_open' in df.columns:
            
            first_bet_open = df.loc[i, 'first_open']
            second_bet_open = df.loc[i, 'second_open']

            if first_bet_open < second_bet_open:
                series_open.append(df.loc[i, 'first_team'])
            else:
                series_open.append(df.loc[i, 'second_team'])
            
            first_bet_close = df.loc[i, 'first_close']
            second_bet_close = df.loc[i, 'second_close']

            if first_bet_close < second_bet_close:
                series_close.append(df.loc[i, 'first_team'])
            else:
                series_close.append(df.loc[i, 'second_team'])
            
                
        else:
            first_bet_open = df.loc[i, 'first_open_spread']
            second_bet_open = df.loc[i, 'second_open_spread']

            if math.isnan(first_bet_open):
                series_open.append(df.loc[i, 'second_team'])
            else:
                series_open.append(df.loc[i, 'first_team'])
                
            
            first_bet_close = df.loc[i, 'first_close_spread']
            second_bet_close = df.loc[i, 'second_close_spread']

            if math.isnan(first_bet_close):
                series_close.append(df.loc[i, 'second_team'])
            else:
                series_close.append(df.loc[i, 'first_team'])
        
    return series_open, series_close
        
```

### Find Winner

Based on the scores given in each dataframe, we calculate who the winner is in each match.

```python
def find_winner(df):
    winner = []
    for i in range(0, len(df)):
        first_score, second_score = df.loc[i, 'score'].split('-')
        
        if int(first_score) < int(second_score):
            winner.append(df.loc[i, 'second_team'])
        elif int(first_score) > int(second_score):
            winner.append(df.loc[i, 'first_team'])
        else:
            winner.append('Draw')
    
    return winner
```

In `compare_columns`, we pass in two columns and see how similar they are. The function compares the values in each column at a given index and appends either a 0 (don't match) or a 1 (do match).

```python
def compare_columns(first_col, second_col):
    values_same = []
    for i in range(0, len(first_col)):
        if first_col[i] == second_col[i]:
            values_same.append(1)
        else:
            values_same.append(0)
    
    return values_same
```

## Predictive Model: Logistic Regression

We are using a logistic regression as we are attempting to classify winners of each match (a binary value).  It wouldn't make sense to use a linear regression, as this is not continuous data.

For each model:
 - We randomly shuffle our rows in our sport dataframe because the playoffs occur in the last 20% of our data, which is all in the test set therefore unfairly skewing the data. As a result, each time the cell is run the precision changes by a small percentage.
 - We select the desired columns as predictors.
     - If the dataframe contains the spread (hockey and baseball), we replace the NaN values with 0.
 - We compare the scores of the two teams and determine the winner (0 if first team won, 1 if second team won).
 - We split our data into training and test sets using 80% of the data for the training set and 20% for the test set.
 - For our logistic regression, we used the default solver (lbfgs).

```python
def classify_winner(row):
    if row[0]==row[2]:
        return 0
    else:
        return 1
```

## Football Predictive Model

For this predictive model, we will be using the following as our predictors:
 - `first_open_spread` : the first team's opening spread, if any
 - `first_close_spread` : the first team's closing spread, if any
 - `first_ML` : the first team's closing money line
 - `second_ML` : the second team's closing money line

Notice, we only use the first team's opening and closing spread since spread is only assigned to one team per  game. We thought about adding the second team's opening and closing spread as the negative of the first team's spread. However, this did not add any value to our predictive model.

```python
shuffle_football_df = new_football_df.sample(frac=1).reset_index(drop=True)
predictive_football_df = shuffle_football_df[['first_open_spread', 'first_close_spread', 'first_ML', 'second_ML']]
predictive_football_df = predictive_football_df.replace(np.nan,0)
predictive_football_df.head()
```

```python
outcome = shuffle_football_df[['first_team','second_team','winner']] 
outcome['winner'] = outcome.apply(classify_winner, axis=1)
outcome.head()
```

```python
num_training = .8 * len(predictive_football_df)

X_train = predictive_football_df.loc[:num_training, :]
y_train = outcome.loc[:num_training, 'winner']
X_test = predictive_football_df.loc[num_training:, :]
y_test = outcome.loc[num_training:, 'winner']
```

```python
log_reg = LogisticRegression(solver='lbfgs')
log_reg.fit(X_train, y_train)
y_pred = log_reg.predict(X_train)
print(classification_report(y_train, y_pred))
```

```python
y_pred = log_reg.predict(X_test)
print(classification_report(y_test, y_pred))
```
