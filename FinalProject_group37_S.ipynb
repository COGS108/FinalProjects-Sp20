### Non-Parametric Test

Since the global sales distributions of ESRB ratings are skewed and non-normal, we should use non-parametric tests. 

First we use the Kruskal-Wallis H-test, a non-parametric version of ANOVA, to test whether the medians of all of the groups are equal. The Kruskal-Wallis H-test assumes that H has a chi square distribution, so the number of samples in each group must not be too small, which our dataset qualifies. 

- H0: The population median of all of the groups are equal.
- H1: The population median of all of the groups are not equal.

```python
# Kruskal-Wallis H-test
statistic, pvalue = stats.kruskal(df_ec['log_Global_Sales'], 
                                  df_e['log_Global_Sales'],
                                  df_e10['log_Global_Sales'],
                                  df_t['log_Global_Sales'],
                                  df_m['log_Global_Sales'],
                                  df_rp['log_Global_Sales'])
statistic, pvalue
```

Such a small p-value suggests to reject the null hypothesis that the medians of all ESRB ratings are equal. To learn which ESRB rating differs from others, we decide to run Dunn's test from the scikit-posthocs package as our non-parametric pairwise multiple comparisons (post hoc) test.

Because the post-hoc test will produce multiple p-values, adjustments to the p-values can be made to avoid inflating the possibility of making a type-I error. Here we use the Holm–Bonferroni method to adjust the rejection criteria for each of the individual hypothesesis by controlling the family-wise error rate (probability that one or more Type I errors will occur).

```python
# post hoc Dunn's test
sp.posthoc_dunn(df_rating, val_col='log_Global_Sales', group_col='ESRB_Rating', p_adjust='holm')
```

The table shows that the p-values for M rating and RP rating are generally smaller than the significance level of 0.05, thus M rating and RP rating differ from the rest groups of the ratings. They are not from the same population.

### Bootstrap Hypothesis Test

Next, we use bootstrap hypothesis test to perform exploratory analysis about the relationship between `log_Global_Sales` and `ESRB_Rating`. In video game industry, M rating or higher is restricted to users who are appropriately aged. We collapse the groups of `ESRB_Rating` into two: the restricted group (M rating) and nonrestricted group (EC, E, E10, T ratings). Since RP rating represents rating pending, we are not sure how to group it, so we do not include it in our bootstrap test. 

- H0: The difference in means between the restricted group and the nonrestricted group is 0.
- H1: The difference in means between the restricted group and the nonrestricted group is greater than 0.

```python
# create a function to perform bootstrap hypothesis test on every input rating category for n rounds
def bootstrap(rating, n):
    sample_data = df_rating[['ESRB_Rating', 'log_Global_Sales']]
    sample_data = sample_data[sample_data['ESRB_Rating'] != 'RP']
    true_data = sample_data[sample_data['ESRB_Rating'] == rating]['log_Global_Sales']
    false_data = sample_data[sample_data['ESRB_Rating'] != rating]['log_Global_Sales']
    obs_stat = np.mean(true_data) - np.mean(false_data)
    
    means = []
    for _ in range (n):
        bootstrap = sample_data.sample(sample_data.shape[0], replace=True)
        true_mean = bootstrap.iloc[:true_data.shape[0]]['log_Global_Sales'].mean()
        false_mean = bootstrap.iloc[true_data.shape[0]:]['log_Global_Sales'].mean()
        means.append(true_mean - false_mean)

    plt.hist(means)
    plt.scatter(obs_stat, 0, color='red')
    plt.title('Distribution of Difference in Means')

    pval = np.mean(pd.Series(means) >= obs_stat)
    print(pval)
```

```python
bootstrap('M', 50000)
```

According to the test result, the shape of our bootstrap distribution is normal, and the p-value is 0 which is lower than the significance level of 0.05. We reject the null hypothesis that the difference in means between the restricted group and the nonrestricted group is 0. Thus we have evidence to conclude that the average `log_Global_Sales` of the restricted group is higher than the average `log_Global_Sales` of the nonrestricted group.

### Conclusion for the Relationship between ESRB Ratings and Total Global Sales

We find there is a relationship between ESRB ratings and total global sales of video games. 

Both the results of the Kruskal-Wallis H-test and the bootstrap hypothesis test show that different ESRB ratings do not come from the same population. 

The Kruskal-Wallis H-test result suggests that some of the medians of the ESRB ratings are statistically significantly different from others. Then the post hoc Dunn's test result demonstrates that M rating and RP rating are the groups differ from the rest ESRB ratings. 

Then we use bootstrap hypothesis test as another way of exploratory analysis. We collapse the ESRB ratings into the restricted group (M rating) and the nonrestricted group (E, EC, E10, T ratings). The result shows that the shape of our bootstrap distribution is normal, and the null hypothesis that the difference in means between the two groups is 0 should be rejected. The average global sales of the restricted group is higher than that of the nonrestricted group.

### Non-Parametric Test

Since the global sales distributions of genres are skewed and non-normal, we should use non-parametric tests. 

First we use the Kruskal-Wallis H-test, a non-parametric version of ANOVA, to test whether the medians of all of the groups are equal. The Kruskal-Wallis H-test assumes that H has a chi square distribution, so the number of samples in each group must not be too small, which our dataset qualifies. 

- H0: The population median of all of the groups are equal.
- H1: The population median of all of the groups are not equal.

```python
# Kruskal-Wallis H-test
statistic, pvalue = stats.kruskal(find_genre('Action'), 
                                  find_genre('Sports'),
                                  find_genre('Misc'),
                                  find_genre('Adventure'),
                                  find_genre('Role-Playing'),
                                  find_genre('Shooter'),
                                  find_genre('Racing'),
                                  find_genre('Simulation'),
                                  find_genre('Platform'),
                                  find_genre('Strategy'),
                                  find_genre('Fighting'),
                                  find_genre('Puzzle'),
                                  find_genre('Action-Adventure'),
                                  find_genre('Music'),
                                  find_genre('Visual Novel'),
                                  find_genre('MMO'),
                                  find_genre('Party'),
                                  find_genre('Sandbox'),
                                  find_genre('Education'),
                                  find_genre('Board Game'))
statistic, pvalue
```

Such a small p-value suggests to reject the null hypothesis that the medians of all genres are equal. To learn which genre differs from the others, we decide to run Dunn's test from the scikit-posthocs package as our non-parametric pairwise multiple comparisons (post hoc) test.

```python
# post hoc Dunn's test
dunn_test = sp.posthoc_dunn(df_genre, val_col='log_Global_Sales', group_col='Genre', p_adjust='holm')
dunn_test
```

To reduce the dimension of the Dunn's test, we decide to remove any genre with more than half of its p-values being non-significant. The remaining genres are those that differ the most from the rest groups of the genres.

```python
sig_dunn_test = dunn_test.copy()

# drop genres with non-significant Dunn's test p-values
to_drop = []
for genre in sig_dunn_test.columns.to_list():
    if sig_dunn_test[sig_dunn_test[genre] > 0.05].shape[0] >= 10:
        to_drop.append(genre)
        
sig_dunn_test = sig_dunn_test.drop(to_drop, axis=1)
sig_dunn_test = sig_dunn_test.drop(to_drop, axis=0)
sig_dunn_test
```

The table shows that the p-values of the 9 genres (Action, Adventure, Misc, Puzzle, Racing, Role-Playing, Simulation, Strategy and Visual Novel) are generally smaller than the significance level of 0.05. They differ from the rest groups of the genres, thus they are not from the same population.

### 2-Sample Kolmogorov-Smirnov Statistics Test

We use the 2-sample ks statistics test to compare the empirical cumulative distribution function of log_Global_Sales of one genre and that of all genres.

We perform the test on each genre and record the test statistics and p-values into a dataframe.

- H0: The two distributions are identical.
- H1: The two distributions are not identical.

```python
statistics = []
pvals = []
dvals = []

# compare log_Global_Sales distribution of each genre with that of all genres
y = df_genre['log_Global_Sales'].values
n2 = df_genre['log_Global_Sales'].shape[0]
for genre in df_genre['Genre'].value_counts().index.to_list():
    x = df_genre[df_genre['Genre'] == genre]['log_Global_Sales'].values
    n1 = df_genre[df_genre['Genre'] == genre].shape[0]
    # the coefficient c(a) for the significance level of 0.05 is 1.36
    d = 1.36 * np.sqrt((n1 + n2) / (n1 * n2))
    stat, pval = stats.ks_2samp(x, y)
    statistics.append(stat)
    pvals.append(pval)
    dvals.append(d)

# record test results and sort by p-values in ascending order
ks_test = pd.DataFrame({'test stats': statistics, 
                        'p-value': pvals, 
                        'critical value': dvals}, 
                       index=df_genre['Genre'].value_counts().index.to_list())
ks_test = ks_test.sort_values(by=['p-value'])
ks_test
```

```python
# select genres with significant test statistics
ks_test[ks_test['test stats'] >= ks_test['critical value']].index.to_list()
```

According to the table above, the 15 genres (Adventure, Sports, Visual Novel, Strategy, Shooter, Platform, Misc, Puzzle, Action-Adventure, Fighting, Action, Role-Playing, Simulation, MMO, Sandbox) have test statistics larger than the calculated critical values. Thus we reject the null hypothesis that the log_Global_Sales of each one of them and that of all genres are from the same continuous distribution. We have enough evidence to conclude that these 15 genres have different distributions than the log_Global_Sales distribution of all genres.

In addition, out of the previous 9 genres as well as these 15 genres, we notice that some of the genres are the relatively popular ones and the others are not. Thus we don't have enough evidence to support our hypothesis that global sales of video games with a popular genre usually stand out from that of all genres.

### Conclusion for the Relationship between Genres and Total Global Sales

We find there is a relationship between genres and total global sales of video games.

Both the results of the Kruskal-Wallis H-test and the 2-sample Kolmogorov-Smirnov statistics test show that different genres do not come from the same population.

The Kruskal-Wallis H-test result suggests that some of the medians of the genres are statistically significantly different from others. Then the post hoc Dunn's test result demonstrates that the 9 genres (Action, Adventure, Misc, Puzzle, Racing, Role-Playing, Simulation, Strategy and Visual Novel) differ from the rest groups of the genres, thus they are not from the same population.

The 2-sample Kolmogorov-Smirnov statistics test result demonstrates that the 15 genres (Adventure, Sports, Visual Novel, Strategy, Shooter, Platform, Misc, Puzzle, Action-Adventure, Fighting, Action, Role-Playing, Simulation, MMO, Sandbox) have different distributions than the log_Global_Sales distribution of all genres. 

Since some of the above genres are the relatively popular ones and the others are not, we don't have enough evidence to support our hypothesis that global sales of video games with a popular genre usually stand out from that of all genres.

### Further Analysis: Is there a confounding variable that affect the correlation between ESRB_Rating and Global_Sales of video games?

According to the previous analysis, we find that there is a stastically significant relationship between ESRB ratings and global sales of video games. We wonder if this relationship is due to the fact that the ESRB rating got by a video game often matches with the target player groups of the video game. For example, if a video game is designed to be played by everyone including children, it will not contain strong violent or sexual contents that would cause it to be classified as restricted (M rating). To analyze the actual effects of ESRB ratings on sales of video games, we want to divide the dataset into two groups, one contains the sales of video games before the video game content rating system was adopted, and one contains the sales of video games after the video game content rating system was adopted. We will compare the differences between the sales distribution of games of the same rating and that of games of all ratings for both groups.

Because different countries have different video game content rating systems that were adopted in different years and with different classification rules, we are not able to use `Global_Sales` in this part. We turn to investigate our question using local sales of video games. 

Originally, we are going to use `ESRB_Rating` and `NA_Sales`, since ESRB is an American organization. However, as ESRB rating system was established and adopted early in 1994, we do not have enough data to analyze the difference of sales before and after the system was adopted. 

Since the video game content rating system of Japan, CERO, was adopted since January 2006, and it has the same restriction and nonrestriction groups as the video game content rating system of the United States, ESRB, we decide to use `JA_Sales` instead of `NA_Sales` to carry out our analysis here.

```python
# starting years of video game content rating systems of different countries
rating_start_year.head()
```

```python
# reload dataset
df_rating_jp = pd.read_csv('vgsales-12-4-2019.csv')
# select necessary columns and drop null values
df_rating_jp = df_rating_jp[['Name', 'ESRB_Rating', 'JP_Sales', 'Year']]
df_rating_jp = df_rating_jp.dropna(axis=0).reset_index(drop=True)
df_rating_jp.shape
```

According to previous test results, restricted rating (M rating) games have sales distribution differ from the rest  groups of the ratings. Thus we want to test if such a difference is consistent before and after the video game content rating system of Japan was adopted. We first filter all M rating games, and split the datasets according to their released `Year`. Then we use boxplots and density plots to compare the distributions of `JP_Sales`.

```python
# filter M rating games
df_rating_jp_m = df_rating_jp[df_rating_jp['ESRB_Rating'] == 'M'].reset_index(drop=True)
df_rating_jp_m.head()
```

```python
# split the datasets into before and after the rating system was adopted
start_year = rating_start_year[rating_start_year['country'] == 'Japan']['start year'].to_list()[0]
df_rating_jp['All Ratings'] = df_rating_jp['Year'].apply(lambda x: 'before' 
                                                          if x < start_year else 'after')
df_rating_jp_m['M Rating'] = df_rating_jp_m['Year'].apply(lambda x: 'before' 
                                                          if x < start_year else 'after')

# examine JP_Sales distributions for all games before vs. after ESRB rating system was adopted 
sns.set(style='whitegrid', font_scale=1.5)
plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 1)
plt.ylim(0, 3)
ax = sns.boxplot(x='All Ratings', y='JP_Sales', data=df_rating_jp, order=['before', 'after'])
ax.axes.set_title('Before ESRB All Ratings JP Sales Distribution', fontsize=25)

# examine JP_Sales distributions for M rating games before vs. after ESRB rating system was adopted 
plt.subplot(1, 2, 2)
plt.ylim(0, 3)
sns.boxplot(x='M Rating', y='JP_Sales', data=df_rating_jp_m, order=['before', 'after'])
ax.axes.set_title('After ESRB M Rating JP Sales Distribution', fontsize=25)
```

Note that the left boxplot shows the JP Sales distribution before and after the video game content rating system was adopted for all games and the right one shows that of M rating games only.

There is no obvious difference between the two boxplots. For better comparison, we plot the kde lines in the same density plot.

```python
df_rating_jp_pre = df_rating_jp[df_rating_jp['Year'] < start_year].reset_index(drop=True)
df_rating_jp_post = df_rating_jp[df_rating_jp['Year'] >= start_year].reset_index(drop=True)
df_rating_jp_m_pre = df_rating_jp_m[df_rating_jp_m['Year'] < start_year].reset_index(drop=True)
df_rating_jp_m_post = df_rating_jp_m[df_rating_jp_m['Year'] >= start_year].reset_index(drop=True)

# examine JP_Sales distributions using a density plot
sns.set(style='whitegrid', font_scale=1.5)
plt.figure(figsize=(20, 10))
sns.distplot(df_rating_jp_m_pre['JP_Sales'], hist=False, color='tab:blue', label='before Rating M rating')
sns.distplot(df_rating_jp_pre['JP_Sales'], hist=False, color='tab:orange', label='before Rating All ratings')
sns.distplot(df_rating_jp_m_post['JP_Sales'], hist=False, color='tab:green', label='after Rating M rating')
sns.distplot(df_rating_jp_post['JP_Sales'], hist=False, color='tab:red',
             label='after Rating All ratings').set_title('Each ESRB Rating Category Global Sales Distribution', 
                                                       fontsize=25)
```

We find out that before the adoption of video game content rating system, the difference between the JP Sales distribution of all games (orange line) and that of M rating games (blue line) is larger.

### 2-Sample Kolmogorov-Smirnov Statistics Test

We use the 2-sample ks statistics test to test whether the JP_Sales of M rating and that of all ratings are from the same continuous distribution. In addition, to better compare the differences between before and after the video game content rating system was adopted, we split the test into two. 

**Test 1: Before the video game content rating system was adopted.**
- H0: JP Sales distribution of all games and that of M rating games are identical.
- H1: JP Sales distribution of all games and that of M rating games are not identical.

```python
# compair JP_Sales distribution of M rating with that of all ratings before the rating system started
x = df_rating_jp_m_pre['JP_Sales'].values
y = df_rating_jp_pre['JP_Sales'].values
n1 = len(x)
n2 = len(y)
d = 1.36 * np.sqrt((n1 + n2) / (n1 * n2))
stat, pval = stats.ks_2samp(x, y)

print(pval)
```

**Test 2: After the video game content rating system was adopted.**
- H0: JP Sales distribution of all games and that of M rating games are identical.
- H1: JP Sales distribution of all games and that of M rating games are not identical.

```python
# Test 2: compair JP_Sales distribution of M rating with that of all ratings since the rating system started
x = df_rating_jp_m_post['JP_Sales'].values
y = df_rating_jp_post['JP_Sales'].values
n1 = len(x)
n2 = len(y)
d = 1.36 * np.sqrt((n1 + n2) / (n1 * n2))
stat, pval = stats.ks_2samp(x, y)

print(pval)
```

According to the result p-values, both tests fail to reject the null hypothesis that the JP_Sales distribution of M rating with that of all ratings are from the same continuous distribution. Thus our conclusion about the relationship between ESRB ratings and global sales does not imply that there is also a relatinoship betwen ESRB ratings and local sales. 

In addition, if it is the ESRB rating system that makes sales of M rating games different than others, the difference between the JS_Sales distribution should be larger, which means the p-value we get from the second test should be smaller than that from the first one. However, the test results show that the second p-value is larger than the first p-value, which indicates that the difference between the JS_Sales distribution of M rating and that of all ratings is smaller after the video game content rating system was adopted in Japan.  

Thus our suspect that there might be a confounding factor, game preferences of players, behind the correlation between ESRB ratings and global sales is validated.
