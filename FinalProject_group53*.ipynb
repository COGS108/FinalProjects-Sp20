Our second dataset from Environmental Impacts of Food Production was already in a tidy data format because it was downloaded from a CSV file. Once the two datasets were merged, our data became more clean because many of the values that were missing from the first dataset were present in the second dataset, so lots of null values were eliminated in the merged dataset. Therefore, once we merged the two datasets, all of the values for each row took on the same metrics, significant figures, and datatypes. This made it easier to use these values in future analyses.

The first dataset from the Meat Eater’s Guide to Climate Change and Health consisted of 21 separate tables, with each table pertaining to a type of food. Each table specified the kg of CO2 equivalent produced in the production of a pound of the given food and the kg of CO2 equivalent produced in the post-production of a pound of that food. In order to initially get this information into a single dataset, we created a Google Spreadsheet with 21 observations, where each observation was one of the food types specified in the initial dataset. We then took the values from the initial dataset and converted them into kg of CO2 per kg of food at production and kg of CO2 per kg of food at post-production. The second dataset from Environmental Impacts of Food Production was already in a usable format because it was a downloadable CSV file.

Our two datasets had data in a usable format once they had been cleaned, but in order to have the data be relevant and related to our project we needed to transform the data from the Meat Eater’s Guide to Climate Change and Health dataset. Initially the features for this dataset were kilogram emissions per kilogram of food pre-farmgate and post-farmgate. When comparing the features included in the two original datasets, we had to transform some of the features from the Meat Eater’s Guide to Climate Change and Health dataset so that they matched the features from the Environmental Impacts of Food Production dataset. The changes made to the dataset can be found from the first link under Sources (which is listed below). We needed to transform this data into kilograms of emissions per 1000 kcal of that food consumed and kilograms of emissions per 100 grams of protein. We made these transformations using Excel functions and nutritional information from the USDA (Source 2, Source 3). The USDA provided data that gave the grams of protein for that food per 100 gram of its weight along with the calories per 100 grams of weight. Utilizing this we set up formulas that took the sum of the emissions for each type of food pre-farmgate and post-farmgate to give total emissions of the food item over its lifetime. We then multiplied that sum by the calculation of how many kilograms of that food would provide 1000 kcal (which was calculated by dividing 1000 kcal by kcal per 100 grams of that food item and then further dividing by 10) to find the kilogram emissions per 1000 kcal for a given type of food. Our second formula took the sum of the emissions over the lifetime of the food and multiplied it by the ratio of 100 grams of weight to grams of protein provided by that food item per 100 grams of weight (this ratio gives the amount of weight needed to provide 1 gram of protein) and further divided this number by 10 to convert it to kilograms, in order to comply with the units provided by the existing dataset to calculate emissions. The auxiliary data calculated was for annual projections of emissions for each food and this was calculated by summing the emissions per kilogram of a given food type and multiplying the sum by its annual consumption in kilograms (Source 4, Source 5, Source 6). Ultimately these calculations provide the principal pieces of data for our project which are greenhouse gas emissions per 1000kcal (kgCO2eq per 1000kcal) and greenhouse gas emissions per 100g protein (kgCO2eq per 100g protein).

Afterwards we compared the data we had compiled to what was needed for the 8 diets. We analyzed what types of food each diet consisted of and subsequently formulated meal plans based on a 2000 daily caloric intake with the appropriate food types and macronutrient ratios for each diet. We made this dataset by investigating 3 different daily meals of the 8 diets, and considering the amount and what types of foods that went into those foods. We then got the nutritional information for those food types and scaled the protein, fat, and carbohydrate values to the amount of food used in the meal to get the calories and grams used for that kind of food in the meal (Source 3). Once we had the generic meal plans configured for each diet we took all of the foods listed for each diet’s meal plan, classified them each as one of the food groups from our merged dataset, and used the calorie and gram counts for each observation to compute the daily kg of CO2 equivalent emissions for the foods included in each diets’ daily meal plan.

### Load datasets containing foods related to the 8 different diets. 

#### Get the information on 8 different diets based on daily meal plans that adhere to each diet's specifications.

```python
# diet data calculations
keto_paleo_data = pd.read_csv('./datasets/diet_data/Paleo_Keto_Diets.csv')
omni_whole_flex = pd.read_csv('./datasets/diet_data/Omnivore_Whole30_Flexitarian_Diets.csv')
vegan_veggie_pesca = pd.read_csv('./datasets/diet_data/Vegan_Vegetarian_Pescatarian.csv')
```

#### Merge these 3 dataframes together to get one complete dataframe for the details of all considered diets.

```python
# put csvs together
diets = keto_paleo_data.merge(omni_whole_flex, how='outer')
diets = diets.merge(vegan_veggie_pesca, how='outer')
diets
```

#### Get the emissions per calories/grams of each food that is considered a part of a daily meal of a diet 

```python
# calculate the emissions for each calorie/gram amount of food
def getFoodEmissions(food, food_cals, food_grams):
    try:
        # handle a couple types of food
        food = food.lower()
        food = food.strip()
        if food == 'butter':
            food = 'milk'
        elif food == 'mayonnaise':
            food = 'eggs'
        elif food == 'coconut oil':
            food = 'nuts'
        elif food == 'avocado':
            food = 'banana'
        elif 'oil' in food:
            food = 'vegetables'
        elif 'coffee' in food:
            # adjust the grams of actual coffee beans since we are considering 1 cup of coffee 
            food_grams = 10
        
        return (pd.Series(
            { 
                'Calorie_Emission': (food_cals / 1000) * emissions.loc[food]['kg_ghg_per_1000Cal'],
                'Gram_Emission': (food_grams / 100) * emissions.loc[food]['kg_ghg_per_100g']
            }
        ))
    except KeyError:
        # shouldn't need this anymore
        print(f"No data on {food}")
        return np.nan, np.nan

diets = pd.concat([diets, diets.apply(lambda row: getFoodEmissions(row['Food'], row['Calories'],                                                                      row['Grams']), axis=1)], axis=1)
diets
```

It appears that the diets with the greatest percentages of protein in their daily calorie intake (Keto, Paleo) have largest CO2 emissions based on the foods that make up their daily meals. This is because proteins, especially meat-based proteins, generally have the greatest CO2 emissions. So, diets that obtain more of their calories from carbohydrates end up having less emissions as a part of their meals. Also, the diets whose protein intake comes from specific meat/non-meat sources (Pescatarian, Vegan, Vegetarian) end up having lower emissions related to their meals.

Therefore, the greenhouse gas emission profiles and environmental impacts of such diets may not be considered in our analysis. We handled this issue by explicitly stating that we are solely investigating the greenhouse gas emissions associated with mainstream diets of our choosing. If further ethical issues had arisen, we had planned to either stop using the problematic datasets, whether they be biased or restricted data, or to erase personally identifiable information.
