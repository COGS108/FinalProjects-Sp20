We believe that population density and air pollution will correlate most strongly with COVID-19 cases and deaths respectively in the United States. The NY Times dataset had multiple entries by date, but the numbers for cases and deaths were cumulative, therefore we only needed the most recent date to get a full count of COVID-19 cases and deaths. We also removed the unecessary decimal place from the fips code.

```python
# Extract most recent data from nytimes dataset and exclude all else
nytimes_county = nytimes_county.loc[nytimes_county['date']=='2020-06-07'].reset_index(drop=True)
nytimes_county = nytimes_county.drop(columns='date')

# Remove decimal place from fips
nytimes_county['fips'] = nytimes_county['fips'].astype(str)
nytimes_county['fips'] = nytimes_county['fips'].str[:-2]

# Take a look
nytimes_county.head()
```

```python
nytimes_county[nytimes_county['county']=='New York City']
```

The NY Times dataset does not break New York City into 5 boroughs each with a respective fips code. The NY Times dataset puts all data into NYC which is listed without a fips code and has no entries for the 5 boroughs of which NYC is composed. This means that the data gets lost when we merge with other dataframes (all of which have data for the boroughs). To solve this problem, we equally allocated all NYC cases and deaths to NYC's five boroughs.

```python
# Averaging NYC cases and deaths
cases_per_bor = nytimes_county[nytimes_county['county']=='New York City'].cases.item() / 5
deaths_per_bor = nytimes_county[nytimes_county['county']=='New York City'].deaths.item() / 5

# Allocating COVID-19 cases and deaths to NYC's 5 boroughs
nytimes_county = nytimes_county.append({'county':'Queens', 'state':'New York', 'fips':'36081', 'cases':cases_per_bor, 'deaths': deaths_per_bor}, ignore_index=True)
nytimes_county = nytimes_county.append({'county':'Bronx', 'state':'New York', 'fips':'36005', 'cases':cases_per_bor, 'deaths': deaths_per_bor}, ignore_index=True)
nytimes_county = nytimes_county.append({'county':'Kings', 'state':'New York', 'fips':'36047', 'cases':cases_per_bor, 'deaths': deaths_per_bor}, ignore_index=True)
nytimes_county = nytimes_county.append({'county':'New York', 'state':'New York', 'fips':'36061', 'cases':cases_per_bor, 'deaths': deaths_per_bor}, ignore_index=True)
nytimes_county = nytimes_county.append({'county':'Richmond', 'state':'New York', 'fips':'36085', 'cases':cases_per_bor, 'deaths': deaths_per_bor}, ignore_index=True)
```

In order to understand the correlation of covid cases with the socioeconomic status of every county in the country, we need to rename to merge our demographic data with the number of cases per county.

```python
#Merge the number of cases, deaths, fips with the us demographic to make a comparison between poverty levels and number of cases.
CasesSocioeconomicGraph = pd.merge(nytimes_county_demo, df_demogr_socioeconomic, sort = True, how = 'outer')
CasesSocioeconomicGraph.head()
```

```python
#Counties that have no cases/deaths/fips will be dropped.
CasesSocioeconomicGraph = CasesSocioeconomicGraph.dropna(subset=['cases', 'deaths', 'fips'])

#Calculate the percentage of cases in relation to the county's population
CasesSocioeconomicGraph['CovidCasesPerPop'] = CasesSocioeconomicGraph['cases']/(CasesSocioeconomicGraph['TotalPop'] )
CasesSocioeconomicGraph['CovidDeathsPerPop'] = CasesSocioeconomicGraph['deaths']/(CasesSocioeconomicGraph['TotalPop'])
CasesSocioeconomicGraph.head()
```

A scatter plot alone is not enough to show to a correlation. Therefore, an OLS is needed to show how income levels and poverty levels affect the Covid cases in each county in the country.

```python
#Scatter plot to show correlation between covid cases and income
scatterCorr = CasesSocioeconomicGraph.plot.scatter(y = 'CovidCasesPerPop', x = 'Income', s = 2, c = '#FF6666')
```

```python
x1 = CasesSocioeconomicGraph['Income']
y1 = CasesSocioeconomicGraph['CovidCasesPerPop']
outcome, predictors = patsy.dmatrices('y1 ~ x1')
mod = sm.OLS(outcome, predictors)
res = mod.fit() 
print(res.summary())
```

In order for an OLS regression to be valid, the residual (error) should be normally distributed. If it is, it should fit to the line of our probability plot. The tail of this graph does diverge significantly.

```python
# Check the distribution of the error
pred_val = res.fittedvalues.copy()
```

```python
fig, ax = plt.subplots(figsize=(6,2.5))
graph = stats.probplot(res.resid, plot=ax, fit=True)
```

The mean of residuals should be approximately zero for OLS regression. Ours is approximately zero, which looks good.

```python
res.resid.mean()
```

In order for an OLS regression to be valid, we also need to test for homoscedasticity. The following graph should have no pattern. This graph doesn't look very random, but we can run another test.

```python
# Check for homoscedacitiy
fig, ax = plt.subplots(figsize=(6,2.5))
ax.scatter(res.resid, pred_val, color='#990000')
```

We can also test for homoscedasticity using the White test. A high LM statistic and high F-statistic indicate that the model is homoscedastic which means our OLS results are valid.

```python
white_test = het_white(res.resid,  res.model.exog)
labels = ['LM Statistic', 'LM-Test p-value', 'F-Statistic', 'F-Test p-value']
print(dict(zip(labels, white_test)))
```

## Population Density and COVID-19

Finally we can analyze population density and COVID-19 cases and deaths to see if there is some relationship or correlation. We start by extracting land area data from our Tiger/Line census dataset.

```python
county_area = usa[['STATEFP', 'COUNTYFP',  'NAME', 'ALAND', 'AWATER']]
county_area.head()
```

Because this dataset has fips codes separated by state and county separately, we need to remove leading zeroes from the state fips and merge it with the county fips so that our data will be compatable with our NY Times COVID-19 dataset.

```python
# Treat our fips codes as strings so that we can perform string manipulation on them
county_area['STATEFP'] = county_area['STATEFP'].astype(str)
county_area['COUNTYFP'] = county_area['COUNTYFP'].astype(str)

# Remove leading 0s from state fips
county_area['STATEFP'] = county_area['STATEFP'].str.lstrip('0')

# Make one long fips and combine ALAND and AWATER to have a single area for counties
county_area['fips'] = county_area['STATEFP'] + county_area['COUNTYFP']
county_area['area'] = county_area['ALAND'] + county_area['AWATER']
```

Our Tiger/Line database file has area listed in square meters. We would prefer to have it in square miles. Below is the conversion.

```python
county_area['area'] = county_area['area'] * 0.00000038610215855
```

```python
county_area = county_area[['NAME', 'fips', 'area']]
county_area = county_area.rename(columns={'NAME':'county'})
```
