We want get the monthly information for each state. So we use the last day of each month as the data for each month.

df_states_month stores the data of the sum of daily total number of cases into monthly total number of cases.

```python
states = df_cases['state'].unique()
df_states_month = pd.DataFrame(columns = ['Jan_case','Jan_death','Feb_case','Feb_death','Mar_case','Mar_death'
                                          ,'Apr_case','Apr_death','May_case','May_death','Jun_case','Jun_death'], index = states)
```

```python
for index in df_states_month.index:
    row = df_states_month[df_states_month.index == index]
    df_state = df_cases[df_cases.state == index]
    jandate = df_state[df_state.date == '2020-01-31']
    febdate = df_state[df_state.date == '2020-02-29']
    mardate = df_state[df_state.date == '2020-03-31']
    aprdate = df_state[df_state.date == '2020-04-30']
    maydate = df_state[df_state.date == '2020-05-31']
    jundate = df_state[df_state.date == '2020-06-05']
    if (jandate.empty == False):
        df_states_month.loc[index,'Jan_case'] = jandate.iloc[0,2]
        df_states_month.loc[index,'Jan_death'] = jandate.iloc[0,3]
    if (febdate.empty == False):
        df_states_month.loc[index,'Feb_case'] = febdate.iloc[0,2]
        df_states_month.loc[index,'Feb_death'] = febdate.iloc[0,3]
    if (mardate.empty == False):
        df_states_month.loc[index,'Mar_case'] = mardate.iloc[0,2]
        df_states_month.loc[index,'Mar_death'] = mardate.iloc[0,3]
    if (aprdate.empty == False):
        df_states_month.loc[index,'Apr_case'] = aprdate.iloc[0,2]
        df_states_month.loc[index,'Apr_death'] = aprdate.iloc[0,3]
    if (maydate.empty == False):
        df_states_month.loc[index,'May_case'] = maydate.iloc[0,2]
        df_states_month.loc[index,'May_death'] = maydate.iloc[0,3]
    if (jundate.empty == False):
        df_states_month.loc[index,'Jun_case'] = jundate.iloc[0,2]
        df_states_month.loc[index,'Jun_death'] = jundate.iloc[0,3]
        
pd.set_option("display.max_rows", 20, "display.max_columns", None)
df_states_month = df_states_month.replace(np.nan,0)
df_states_month
```

#### Regression Slope Test for Consumer Sentiment

Consumer Sentiment is the measure used by the Federal Reserve to determine the level of confidence in the American population. We have the prior assumption that consumer sentiment is closely linked to expectations of future income increases.


Stats:
- Average of samples regarding the expected percentage increase of annual income in the United States
- Average of samples regarding consumer sentiment in the United States
- Null $\mu$: $\beta $ = 0
- Alt $\mu$: $\beta $ > 0

```python
#df_cs
income_yearmean
income_cs = pd.concat([df_cs, income_yearmean])
income_cs = income_cs.T
outcome_1, predictors_1 = patsy.dmatrices('MeanIncomeYear ~ UMCSENT', income_cs)
mod_1 = sm.OLS(outcome_1, predictors_1)
res_1 = mod_1.fit()
print(res_1.summary())
income_cs
```

#### Results

Here, we can see in the summary of the model, the fit slope is positive (between 0.023 and 0.04). Assuming the Null hypothesis, the T value of getting the regression slope value of 0.0347 is 6.194. The one sided p-value of 6.194 is 0.00/2 == 0. Regardless of the alpha we choose, the we reject the null hypothesis ($\mu$: $\beta $ = 0) in favor of the alternative hyothesis that the slope is indeed positive. Also, as the r adjusted R-squared value is 0.73, we can see that the regression fits fairly well. Therefore, we have established a positive relationship between the expected increase in annual income and consumer confidence

Then we create a model mathematically describes the relationship between the data points.

Household income has a positive correlation with consumer sentiment. Household income increases when consumer sentiment increases.
