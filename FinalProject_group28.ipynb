# Data Cleaning

In this section, we will read data into a dataframe and apply some data cleaning techniques so that the resulting dataframe only conatins information that is reasonable and necessary in a comprehensible format.

We first read in the whole income dataset.

```python
df = pd.read_csv('./data/2018ACS.csv')
# Check the shape of resulting dataframe
print(df.shape)
# Take a look at the dataset.
df.head()
```

We drop attributes that are not helpful to our research question. 'RACED' divides race into very detailed categories than necessary; therefore, we keep 'RACE', which gives us enough information on the race of the person. 'EDUC' is a general version of education, which only contains whether a person has attended college; therefore, we keep 'EDUCD', which gives us more information on the degree completed and years spent in college.

```python
df = df.drop(['RACED', 'EDUC'], axis=1)
```

We rename 'PWTYPE' to 'City_Status' for readability.

```python
df.rename(columns={'PWTYPE':'City_Status'}, inplace=True)
df.head()
```

Since our research question focuses on the State of California, we drop all observations that are not in California. According to the coding standard, which can be referenced in 'data/2018ACS_ref.pdf', value 06 represents California. Therefore, we drop all rows that are not equal to 6 under attribute 'PWSTATE2'. After filtering states, we can now remove the unnecessary attribute 'PWSTATE2' since our current dataframe only contains observations in California. 

```python
df.drop(df[df['PWSTATE2'] != 6].index, inplace = True) 
df = df.drop(['PWSTATE2'], axis=1)
# Check the shape of resulting dataframe
df.shape
```

According to the INCWAGE Specific Variable Codes, we know that value 999999 represents N/A and 999998 represents missing under the attribute 'INCWAGE'. Therefore, we remove observations that contain invalid wages since it does not give us any information on the salary of the person, which is a crucial factor to our research questions.

```python
df.drop(df[df['INCWAGE'] == 999998].index, inplace = True) 
df.drop(df[df['INCWAGE'] == 999999].index, inplace = True) 
```

Let's have an overview on the distribution of income in our current dataframe.

```python
print(df.describe()['INCWAGE'])
# Check the shape of resulting dataframe
df.shape
```

Notice that the minimum wage is \\$0, which is unreasonable. According to California's Employment Development Department, the minimum weekly benefit for each unemployeed person is \\$40 (Reference: https://edd.ca.gov/Unemployment/UI-Calculator.htm). Since there are 52 weeks in a year, assuming the person is unemployed for the entire year, the minimum yearly benfit for each unemployeed person 40 * 52 = 2,080 dollars. Therefore, we drop all wages below \\$2,080. 

```python
df.drop(df[df['INCWAGE'] < 2080].index, inplace = True) 
```

We now check the distribution of income again after removing unreasonable observations for wages.

```python
print(df.describe()['INCWAGE'])
# Check the shape of resulting dataframe
df.shape
```

Notice that the mean is now 66,070.324833, which agrees with the statistics of the average income in California (Reference: https://smartasset.com/retirement/average-salary-in-california).

We only consider ages between 14(inclusive) and 66(inclusive) because California's minimum age to work is 14 and the retirement age is 66.<br>
References: 
- https://www.employmentlawhandbook.com/wage-and-hour-laws/state-wage-and-hour-laws/california/child-labor-laws/

```python
df.drop(df[df['AGE'] < 14].index, inplace = True) 
df.drop(df[df['AGE'] > 66].index, inplace = True) 
df.reset_index(drop=True, inplace = True)
# Check the shape of resulting dataframe
df.shape
```

## Standardize Attributes

In this section, we will standardize attributes for readability.

### Sex

Since the original labelling for sex uses 1 for male and 2 for female. For readability, we relabel male as 'M' and relabel female as 'F'.

```python
def convert_sex(label):
    return 'M' if label == 1 else 'F'
df['SEX'] = df['SEX'].apply(convert_sex)
# Count in each category
df['SEX'].value_counts()
```

### Race

According to the coding standard, which can be referenced in 'data/2018ACS_ref.pdf', we have the following schema for race:
- 1: White
- 2: Black/African American
- 3: American Indian or Alaska Native
- 4: Chinese
- 5: Japanese
- 6: Other Asian or Pacific Islander
- 7: Other race, nec
- 8: Two major races
- 9: Three or more major races

Since code will be hard to intrepret when doing analysis, we will rename each observation to their corresponding race for readability.

```python
def standardize_race(label):
    race_dict = {1:'White', 
                 2:'Black', 
                 3:'Native American', 
                 4:'Chinese', 
                 5:'Japanese', 
                 6:'Pacific Islander',
                 7:'Other',
                 8:'Two major races', 
                 9:'Three or more major races',
                }
    return race_dict[label]
df['RACE'] = df['RACE'].apply(standardize_race)
# Count in each category
df['RACE'].value_counts()
```

### Education 

After referencing the coding standard, which can be referenced in 'data/2018ACS_ref.pdf', we decided to divide eudcation into a few categories. Following are the categories that we have generalized for each observation:

- No high school (HS)
- Some high school (HS) or General Education Development (GED)
- Associate's degree or lower
- Bachelor‘s Degree
- Master’s Degree
- Professional Degree beyond a Bachelor's degree
- Doctoral Degree

```python
def convert_EDUCD(code):
    if (code <= 26):
        return "No HS"
    elif (code <= 64):
        return "Some HS or GED"
    elif (code <= 83):
        return "Associate's degree or lower"
    elif (code == 101):
        return "Bachelor‘s Degree"
    elif (code == 114):
        return "Master’s Degree"
    elif (code == 115):
        return "Professional Degree beyond Bachelor's"
    elif (code == 116):
        return "Doctoral Degree"
    else:
        return code
df['EDUCD'] = df['EDUCD'].apply(convert_EDUCD)
# Count in each category
df['EDUCD'].value_counts()
```

### Occupation 

We created a new occ_category_new.csv file, which contains a two-column mapping from occupation codes to job types. 
The types of workers are created based on the U.S. Bureau of Labor Statistics's sub-categories found at https://www.bls.gov/soc/2018/major_groups.htm

```python
with open('./data/occ_category_new.csv', mode='r') as infile:
    reader = csv.reader(infile)
    with open('occ_category_write.csv', mode='w') as outfile:
        writer = csv.writer(outfile)
        mydict = {rows[0]:rows[1] for rows in reader}
        
df['Job_Type'] = df['OCC']
df.Job_Type = df.Job_Type.astype(str)
df['Job_Type'] = df['Job_Type'].map(mydict)
print ("number of unique jobs: " + str(len(df['OCC'].unique())))
df = df.drop(['OCC'], axis=1)
print("number of categories: " + str(len(df['Job_Type'].unique())))
```

After cleaning occupation attribute, we converted 529 unique jobs into 98 categories.

### City Status

Since there are many categories in city status, we first check the unique values for city status in our current dataframe before relabelling.

```python
df['City_Status'].unique()
```

Since we only have four unique values for city status, we will rename those values to their corresponding label for each observation so that it will be easier to interpret when we perform analysis in the following sections. According to the coding standard, which can be referenced in 'data/2018ACS_ref.pdf', we have the following schema for city status:
- Code 1: In metropolitan area: In central/principal city
- Code 4: In metropolitan area: Not in central/principal city
- Code 5: In metropolitan area: Central/principal city status indeterminable
- Code 7: Not in metropolitan area

```python
def convert_PWTYPE(code):
    if (code == 1):
        return "principal city"
    elif (code == 4):
        return "not principal city"
    elif (code == 5):
        return "principal city status indeterminable"
    elif (code == 7):
        return "not metropolitan"
    else:
        return code
df['City_Status'] = df['City_Status'].apply(convert_PWTYPE)
# Count in each category
df['City_Status'].value_counts()
```

## Controllable vs. Uncontrollable Factor Analysis

We want to examine the effect of each attribute to income. Since age is a numerical attribute, we would use linear regression to find the Pearson correlation between age and income. However, the rest of attributes are all categorical. Thus, we would use ANOVA to get the eta squared to get the correlation ratio.

```python
# Code obtained from https://www.pythonfordatascience.org/anova-python/
# Function to produce an AVONA table
def anova_table(aov):
    aov['mean_sq'] = aov[:]['sum_sq']/aov[:]['df']
    aov['eta_sq'] = aov[:-1]['sum_sq']/sum(aov['sum_sq'])
    aov['omega_sq'] = (aov[:-1]['sum_sq']-(aov[:-1]['df']*aov['mean_sq'][-1]))/(sum(aov['sum_sq'])+aov['mean_sq'][-1])
    cols = ['sum_sq', 'df', 'mean_sq', 'F', 'PR(>F)', 'eta_sq', 'omega_sq']
    aov = aov[cols]
    return aov
```

# Ethics & Privacy

The main datasets that we are planning to use comes from the Integrated Public Use Microdata Series (IPUMS). We use their services for our own internal, personal, non-commercial use, and not on behalf of or for the benefit of any third party. The dataset can be publicly viewed and downloaded through IPUMS' website.

Ethicality was important to us as we proceeded with our project. We advanced with our analysis while adopting ethical practices centered on removing bias and maintaining privacy. In general, there shouldn't be any equitable impact from our analysis. In regards to privacy, we treat names from observations anonymously, and we filter out observations’ data that are under 14 years old because according to the Fair Labor Standards Act (FLSA), it is the minimum age of employment. Our analysis takes in great consideration in protecting the individuals.

Unfortunately we cannot tell if the data is collected accurately or not. People may report fake salary data which is not a factor in our control, but we will try to collect a larger amount of data samples to avoid this issue.

Our data should be mainly faced to individuals rather than organizations. One potential ethical issue is that some employers may think that our data is good for them to standardized employees’ salary and may make adjustments to their employees’ salary standards. This could harm some employees who have good potential soft-skills that are not listed in our data such as motivation skill, communication skills, or social skills. We will mark our data as “Data for reference use only, and does not represent all job seekers” to communicate the uncertainty in our predictions to those that use our data.

One collection bias factor of our data is all data we have used are from a single website(IPUMS). However, we have fully survey the website to make sure that all data all collected ramdomly, so their should no bias in our data.

Our data ethics is concerned with the following principles:
- **Consent**: If an individual or legal entity would like to use personal data, one needs informed and explicitly expressed consent of what personal data moves to whom, when, and for what purpose from the subject of the data. The subject of the information has the right to know how their data has been used.
- **Privacy**: If data transactions occur all reasonable effort needs to be made to preserve privacy.
- **Openness**: Our data is open to all individual or groups. The data should be freely available and should not have restrictions that would prohibit its use, such as copyright laws
